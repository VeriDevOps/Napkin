```python
def plot_signal(S,offset=0):
    #---------------------------------------------------------------------------------------------------
    def norm(A,offset=0):
    #---------------------------------------------------------------------------------------------------    
        max_a = max(A)
        if max_a == 0:
            return A
        else:
            return [(a/max_a*10) + offset*10.1 for a in A]
    X1 = S['xAxis']
    V1 = S['values']
    plt.step(x=X1,y=np.array(norm(V1,offset)),marker='',where='post',lw=1,alpha=0.5)
               
#---------------------------------------------------------------
def shift_signal(S,dt,from_original_timebase = True):
#---------------------------------------------------------------
    '''
    S  - Signal - {'xAxis':[],'values':[]}
    dt - Float  - dt is added to each sample point in xAxis
    from_original_timebase - Boolean - 
         if True, subsequent shifts are added together
         if False, all previous shifts are reset before
         
    
    '''
   
    ct = S.get('current_shift',0)
    if not from_original_timebase:
        print(f"Continuing from x+{ct} and adding {rt} to each sample")
        newbase = ct + dt
        ct = 0   # Code below first rolls back old shift.
    else:
        print(f"Resetting old {ct} and adding {dt-ct} to each sample")
        newbase = dt # whatever in ct will be rolled back.
    S['xAxis'] = [x - ct + dt for x in S['xAxis']]
    S['current_shift']  = newbase
   
#---------------------------------------------------------------
def unshift_signal(S,dt):
#---------------------------------------------------------------
    '''
        unshift_signal - undo any number of shift_signal operations
    '''
    ct = S.get('current_shift',0)
    S['xAxis'] = [x + dt for x in S['xAxis']]
    S['current_shift'] = 0
```
